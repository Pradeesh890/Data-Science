```python
page = requests.get('https://www.amazon.in/Apple-iPhone-13-128GB-Midnight/product-reviews/B09G9HD6PD/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber=1')
```

```python
page.content
```

```python
import bs4
from bs4 import BeautifulSoup as bs
```

```python
# parsing GET resonse into html page
soup = bs(page.content, 'html.parser')
soup
```

```python
print(soup.prettify())   # pretiffy() converts into formatted HTML page
```

```python
names = soup.find_all('span', class_ = 'a-profile-name')
names
```

```python
customer_name = []
for i in range(len(names)):
    customer_name.append(names[i].get_text())
```
