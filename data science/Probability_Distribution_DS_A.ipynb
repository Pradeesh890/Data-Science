###### Probability Distribution Experiment

```python
# Throwing a fair die multiple times
dice = [x for x in range(1,7)]
dice
```

```python
# defualt num of trials = 10
def throw_dice(n = 10):
  r = []     # list of random results
  for _ in range(n):
      value = random.choice(dice)
      r.append(value)
  return np.array(r)   # function returns array results
```

```python
# call a function to return 10 trails outcome
throw_dice()
```

```python
throw_dice(n = 20)
```

```python
for i in [1,5,10,50,100,500,1000,5000,10000]:
    print("Average of {} dice throws are: {}".format(i,round(throw_dice(i).mean(),2)))
```

###### Bernoulli with Multiple trails.

```python
# This experiment is when a 10 coins are tossed for n number trails
N_trails = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000]
prob = 0.5
average = []
for i in N_trails:
    for n in range(1, i+1):
        average.append(np.mean(bernoulli.rvs(p = prob, size = 10)))
    if (i==10):
        plt.title("Distribution with {} trials of 10 coin tosses".format(i))
        # bins = number of intervals, bins = 10
        plt.hist(average,bins=10,edgecolor='deeppink',color='pink')
        plt.xlim(0.0,1.0)
        plt.xticks([0.1*i for i in range(11)])
        plt.grid(True)
        plt.show()
    else:
        plt.title("Distribution with {} trials of 10 coin tosses".format(i))
        # bins = number of intervals, bins = 25
        plt.hist(average,bins=25,edgecolor='deeppink',color='pink')
        plt.xlim(0.0,1.0)
        plt.xticks([0.1*i for i in range(11)])
        plt.grid(True)
        plt.show()
```

###### Poisson Distribution

```python
from scipy.stats import poisson
```

```python
#  = Average number of events in an interval
```

```python
# Python code to generate PMF using poisson distribution
import pandas as pd

la = 0.5
rv = poisson(la)
x = np.arange(0, 20, 1)
pmf1 = rv.pmf(x)

la = 1
rv = poisson(la)
pmf2 = rv.pmf(x)

la = 5
rv = poisson(la)
pmf3 = rv.pmf(x)

pmf_df1 = pd.DataFrame()
pmf_df1['X'] = x
pmf_df1['PMF1'] = pmf1
pmf_df1['PMF2'] = pmf2
pmf_df1['PMF3'] = pmf3
```

```python
plt.plot(pmf_df1['X'], pmf_df1['PMF1'])
plt.plot(pmf_df1['X'], pmf_df1['PMF2'])
plt.plot(pmf_df1['X'], pmf_df1['PMF3'])
plt.show()
```

### Uniform (continuous) distribution

This is the distribution of the likelihood of uniformly randomly selecting an item out of a finite collection.
For the continuous case, the PDF looks deceptively simple, but the concept is subtle,

$$ f(x)={\begin{cases}{\frac {1}{b-a}}&\mathrm {for} \ a\leq x\leq b,\\[8pt]0&\mathrm {for} \ x<a\ \mathrm {or} \ x>b\end{cases}} $$

```python
from scipy.stats import uniform
```

##### Generate random continious variable (between 0 and 1)

```python
uniform.rvs(size = 6)
```

###### Change *loc* and *scale* parameters to move the range.

```python
# range will random floats between -10 and 0
uniform.rvs(loc = -10, scale = 10, size = 6)
```

```python
# range will random floats between -10 and 10
# range will be -30 to -30+scale = -10, size = number of values
uniform.rvs(loc = -30, scale = 20, size = 6)
```

### Normal (Gaussian) distribution

  * In probability theory, the normal (or Gaussian) distribution is a very common continuous probability distribution. Normal distributions are important in statistics and are often used in the natural and social sciences to represent real-valued random variables whose distributions are not known.

```python
from scipy.stats import norm
# linespace() - generates equally spaced data
x = np.linspace(-3, 3, num = 100)
constant = 1.0 / np.sqrt(2 * np.pi)
# assuming standard normal distribution , stddev = 1, mean = 0
pdf_normal_dist = constant * np.exp((-x ** 2))/2
```

```python
sns.set_style('white')
plt.plot(x, pdf_normal_dist)
plt.title('Normal Distribution')
plt.ylabel('Probability Distribution')
plt.show()
```
